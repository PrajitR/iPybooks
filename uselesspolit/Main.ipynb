# Robotic Politicians

Our government is extremely divided over partisan lines. Republicans and Democrats seem to be perfectly split into two categories. To further examine this divide, I want to start with investigating these questions:

+ How many politicians can be replaced by robots (that is, a robot voting purely on party lines would have the same exact votes as a politician)? 
+ Is there a difference between the House and the Senate? 
+ Is there a difference between Republicans and Democrats? 
+ Are robotic politicians more ideologically extreme?
+ Has there been a trend over the past few years towards more robotic politicians?

**My hypothesis**:

+ A third of the politicians serving in Congress would be considered robotic politicians. There will be more non-robotic politicians than robotic politicians because of all the deals being made.
+ While the House seems to be more ideologically divided, the fact that the House has 435 members increases the likelihood of non-robotic politicians (a lot more deals must be brokered to get a majority).
+ I think the minority party in both the House and the Senate will have a lesser percentage of robotic politicians, because they might be forced to concede on some bills (against their party's line) to gain support on their other bills.
+ I think robotic politicians will be more ideologically extreme because these are the politicians who refuse to compromise on their "values".
+ I definitely think there is a trend towards more robotic politicians because Congress has become increasingly partisan.

```python
def get_bill(congress, year, house, bill):
    """
    Function
    --------
    Gets a single bill from govtrack.us
    
    Parameters
    ---------
    congress : int
        The Congress is the bill in (ex. 113)
    year : int
        The year is the bill in (ex. 2013)
    house : string
        Which Congressional House? ('h' for HoR and 's' for Senate)
    bill : int
        The bill number (ex. 10)
        
    Example
    -------
    get_bill(112, 2011, 'h', 109) 
    """
    url = 'https://www.govtrack.us/data/congress/%d/votes/%d/%s%d/data.json' % (congress,year,house,bill)
    page = requests.get(url).text
    return json.loads(page)

#get_bill(113, 2013, 'h', 109)
```

```python
def congress_from_year(yr):
    """
    Calculates the congress number from a year
    
    congress_from_year(2013)
    > 113
    """
    return int((yr - 1776) / 2) - 6 + (yr % 2)

def get_bills_of_year(year):
    """
    Gets all the bills from a given year. Returns a dictionary
    with two main keys: 'senate' and 'hor'
    """
    
    congress = congress_from_year(year)
    url = 'https://www.govtrack.us/data/congress/%d/votes/%d/' % (congress,year)
    dom = web.Element(requests.get(url).text)
    
    def get_house_bills(house):
        #adapted from CS109 HW5
        """
        Pass in 'h' or 's' to get the house's bills
        """
        votes = [a.attr['href'] for a in dom.by_tag('a') 
             if a.attr.get('href', '').startswith(house)]
        n_votes = len(votes)
        
        # ugh, should have just parsed the links myself instead of this approach 
        
        # there doesn't seem to be a 2011 h484. skips from h483 to h485
        # there doesn't seem to be a 2012 h327. skips from h326 to h328
        def bill_exception(year, bill):
            return ((i == 484) and (year == 2011)) or ((i == 327) and (year == 2012))
        return [get_bill(congress, year, house, i) for i in range(1, n_votes + 1) if not bill_exception(year, i)]
    
    return {'senate': get_house_bills('s'), 'house': get_house_bills('h')}
```

Interestingly, there seem to be continuous up and down cycle for both houses of Congress: for example, in 2007, the House looked at over 1200 bills, then in 2008, they looked at only 800 bills, then in 2009 they looked at over 1000 bills. A good hypothesis would be that politicians have to spend a significant portion of their time campaigning in their election years. This would also explain why the House has a bigger drop than the Senate: *all* House members are up for re-election every two years, while only a $\frac{1}{3}$ of the Senate is up for re-election. I'll check this with hard evidence (do the amount of bills deliberated drop as the election nears?).

```python
# what months are the most bills deliberated on?
senate_months = np.zeros(12)
house_months = np.zeros(12)
for yr in yearrange:
    filename = 'votes%d.json' % yr
    with open(filename) as openfile:
        votes = json.load(openfile)
    for bill in votes['senate']:
        # date format: "2012-01-17T18:53:00-05:00"
        month = int(bill['date'].split('-')[1])
        senate_months[month-1] += 1
    for bill in votes['house']:
        # date format: "2012-01-17T18:53:00-05:00"
        month = int(bill['date'].split('-')[1])
        house_months[month-1] += 1
```

```python
print "Senate: ", senate_months
print "House: ", house_months

norm_senate_months = senate_months / sum(senate_months)
norm_house_months = house_months / sum(house_months)

fig = plt.figure()
ax = fig.add_subplot(111)
width = 0.35
ax.bar(np.arange(12), norm_senate_months, width, label="Senate", color='g', edgecolor='white')
ax.bar(np.arange(12)+width, norm_house_months, width, label="House", color='b', edgecolor='white')
ax.legend(loc="upper left", frameon=False)
plt.title("Month vs Percentage of Bills Deliberated")
plt.xlabel("Month")
plt.ylabel("Percentage of Bills Deliberated")
plt.xticks(np.arange(12)+0.25, ('Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'), rotation=60)
remove_border()
```

For both houses, there is an approximate unimodal distribution, starting and ending an year low, and peaking in June and July. There is a precipitous drop in the month of August, reflecting Congress' month-long August recess. However, there is a also a drop in April. It turns out that Congress has an 11 day recess in April, whereas most months have just a 4 day recess.

Why there is a unimodal distribution, I'm not too sure. Perhaps more influential bills are passed during the summer, which in turn lead to a lot of legislation. That's a question worth investigating in the future.

```python
def partisan_divide(bill, house):
    sides = (['Yea', 'Nay'] if house == 's' else ['Aye', 'No'])
    vote_cast = {k:defaultdict(int) for k in sides}
    for side in sides:
        if side in bill['votes'].keys():
            for v in bill['votes'][side]:
                if v == 'VP': continue # Vice-President Vote
                vote_cast[side][v['party']] += 1
                
    # calculate which side is Democrat and which side is Republican
    def partisanship(vote):
        # closer a threshold is to .5, the less likely a bill will be classified as bipartisan
        # would be a good idea to play around with threshold value
        D_thresh, R_thresh = .8, .2
        # some sides have zero votes.
        if (vote['D'] + vote['R']) == 0: return 'B'
        per = 1. * vote['D'] / (vote['D'] + vote['R'])
        if per >= D_thresh: return 'D'
        elif per <= R_thresh: return 'R'
        else: return 'B'
    
    return {k:partisanship(vote_cast[k]) for k in sides}
```

```python
def robotic_politician(bills, house):
    sides = (['Yea', 'Nay'] if house == 's' else ['Aye', 'No'])
    polit = defaultdict(int)
    for bill in bills:
        partisan = partisan_divide(bill, house)
        for side in sides:
            if side in bill['votes'].keys():
                for v in bill['votes'][side]:
                    if v == 'VP': continue # Vice-President Vote
                    if partisan[side] != 'B' and v['party'] != partisan[side]:
                        polit[v['display_name']] += 1
                    else:
                        # I need all politicians' names - ugly hack...
                        polit[v['display_name']] += 0
    return polit
```

```python
robots_years = []
robot_dem = []
robot_rep = []
for yr in yearrange:
    filename = 'votes%d.json' % yr
    with open(filename) as openfile:
        votes = json.load(openfile)
    robot_senate = robotic_politician(votes['senate'], 's')
    true_robot = [k for k,v in robot_senate.items() if v == 0]
    true_robot_dem = [k for k in true_robot if '(D-' in k]
    robots_years.append(len(true_robot))
    robot_dem.append(len(true_robot_dem))
    robot_rep.append(len(true_robot) - len(true_robot_dem))
```
